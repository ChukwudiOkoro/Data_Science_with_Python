# Sort a dataframe
In pandas, you sort a DataFrame using sort_values().
🔹 Basic syntax.

+ df.sort_values(by, ascending=True, inplace=False)
+ by → column name(s) to sort by
+ ascending → True (default) or False
+ inplace → if True, modifies the DataFrame directly

```python
import pandas as pd
```

```python
# reading the csv file
df_exams = pd.read_csv('Studentsperformance.csv')
```

### 1. Sort by one column (ascending)

```python
df_exams.sort_values(by="math score")
```

###  2. Sort by one column (descending)

```python
df_exams.sort_values(by="math score", ascending=False)  
```

###  3. Sort by multiple columns

```python
df_exams.sort_values(by=["math score", "reading score"], ascending=False)
```

###  4. Different order per column

```python
# sorting by math score in ascending order and reading score in descending order
df_exams.sort_values(by=["math score", "reading score"], ascending=[True, False])
```

### 5. Sort and update original DataFrame

```python
# inplace sorting = True will change the original dataframe without creating a new one
df_exams.sort_values(by=["math score", "reading score"], ascending=False, inplace=True)
df_exams
```

### 6. sort descending with a key function

- lambda x — Creates an anonymous function; x represents the values in the column
- x.str.split() — Splits each string by whitespace (e.g., "Group A" → ["Group", "A"])
- .str[1] — Extracts the 2nd element (index 1) from the split result (e.g., "A")
- .map({...}) — Maps categories to numbers for sorting:
- 'A' → 1
- 'B' → 2
- 'C' → 3
- 'D' → 4
- 'E' → 5
Example:
If your "race/ethnicity" column contains values like "Group A", "Group E", "Group C", this code will:

+ Extract the letter (A, E, C)
+ Convert to numbers (1, 5, 3)
+ Sort by those numbers → A, C, E order

```python
df_exams.sort_values(by="race/ethnicity", ascending=True, key=lambda x: x.str.split().str[1].map({'A': 1, 'B': 2, 'C': 3, 'D': 4, 'E': 5}))
```

```python
# key = lambda and str.lower() function is used to sort the values based simple alphabetical sorting without worrying about uppercase/lowercase differences
df_exams.sort_values(by="race/ethnicity", ascending=True, key=lambda col:col.str.lower())
```
