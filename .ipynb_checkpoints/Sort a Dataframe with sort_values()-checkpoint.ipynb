# Sort a dataframe
In pandas, you sort a DataFrame using sort_values().
🔹 Basic syntax.

+ df.sort_values(by, ascending=True, inplace=False)
+ by → column name(s) to sort by
+ ascending → True (default) or False
+ inplace → if True, modifies the DataFrame directly

```python
import pandas as pd
```

```python
# reading the csv file
df_exams = pd.read_csv('Studentsperformance.csv')
```

### 1. Sort by one column (ascending)

```python
df_exams.sort_values(by="math score")
```
