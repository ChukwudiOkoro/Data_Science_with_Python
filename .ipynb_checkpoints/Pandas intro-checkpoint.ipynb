```python
import pandas as pd
import numpy as np
```

# 1 creating a dataframe from an array

### 1.1 Option 1 with numpy

```python
#creating an array
data = np.array([[1,4], [2,25], [3,6]])
```

# create a dataframe
df  = pd.DataFrame(data, index= ['row1', 'row2', 'row3'],
             columns= ['col1','col2'])

```python
#  showing  the dataframe       #df is a common way to label a DataFrame 
df
```

### 1.2  Option 2 with list shape

```python
#   creating an array with list shape
data =[[1,4],[2,5],[3,6]]
```

```python
# creating a dataframe
df  = pd.DataFrame(data, index= ['row1', 'row2', 'row3'],
             columns= ['col1','col2'])
```

```python
# showing the dataframe
df
```

# 2 Creating DataFrame from a dictionary

```python
#lists used for this example
states = ['Enugu', 'Texas', 'Florida', 'New York']
population = [48489940,  28399408,29299499,19998390]
```

```python
# Storing lists within a dictionary
dict_states = {'States' : states, 'Population' : population}
```

```python
# creating the dataframe
df_population = pd.DataFrame(dict_states)
```

```python
#  showing  the dataframe
df_population
```
