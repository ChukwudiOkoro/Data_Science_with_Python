# Understanding concat() - Combining DataFrames
concat() combines multiple DataFrames into one. The direction depends on the axis parameter:

### Vertical Concatenation (axis=0 or axis='index')
Stacks DataFrames on top of each other — adds rows

````python
import pandas as pd

# Create sample dataframes
df1 = pd.DataFrame({'Name': ['Alice', 'Bob'], 'Age': [25, 30]})
df2 = pd.DataFrame({'Name': ['Charlie', 'Diana'], 'Age': [35, 28]})

print("df1:")
print(df1)
print("\ndf2:")
print(df2)

# Vertical concatenation (stack them)
result = pd.concat([df1, df2], axis=0)
print("\nVertical concat (axis=0):")
print(result)
````

### Horizontal Concatenation (axis=1 or axis='columns')
Places DataFrames side by side — adds columns

````python
# Create sample dataframes
df1 = pd.DataFrame({'Name': ['Alice', 'Bob'], 'Age': [25, 30]})
df3 = pd.DataFrame({'City': ['NYC', 'LA'], 'Salary': [50000, 60000]})

print("df1:")
print(df1)
print("\ndf3:")
print(df3)

# Horizontal concatenation (side by side)
result = pd.concat([df1, df3], axis=1)
print("\nHorizontal concat (axis=1):")
print(result)
````
