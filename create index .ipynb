# create index

```python
import numpy as np
import random
import pandas as pd
```

```python
# reading the csv file
df_exams = pd.read_csv('Studentsperformance.csv')
```

```python
# create non repetitive values for the index
new_index = np.arange(0,1000)
```

```python
# shuffling the indexes
random.shuffle(new_index)
  
```

```python
# create a new column with  new index 
df_exams['new_index'] = new_index
df_exams.head()
```

# set index

inplace=False (default)
→ returns a new sorted/modified DataFrame
→ original DataFrame stays unchanged
inplace=True
→ modifies the original DataFrame directly
→ returns None (this is important!)
When to use it?
+ Use inplace=True → when you don’t need the original data anymore
+ Otherwise → safer to keep default (False)
###### Commom mistake
+  df = df.sort_values(by="age", inplace=True)

👉 This makes df = None ❌
(because the method returns nothing)

```python
# set new index column as index  # inplace : bool, default False
# Whether to modify the DataFrame rather than creating a new one.

df_exams.set_index('new_index', inplace=True) 
df_exams
```

```python
# sort descending by index
df_exams.sort_index(ascending=False)
```

```python
# sort descending by index
df_exams.sort_index(ascending=False, inplace=True)
df_exams
```
