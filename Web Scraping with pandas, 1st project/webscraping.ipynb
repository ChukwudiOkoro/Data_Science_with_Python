```python
# creating a list of leagues
leagues = ["E0", "E1", "E2", "E3", "EC"]

# frame_list will hold one DataFrame per league, appended in the same order
# as `leagues`. That means:
#   frame_list[0] -> data for leagues[0] ("E0")
#   frame_list[1] -> data for leagues[1] ("E1")
# and so on.

frame_list = []   # create an empty list to store the dataframes

for league in leagues:
    url = root + "2526/" + league + ".csv"
    df = pd.read_csv(url)
    # label the frame for clarity (prevents guessing by index later)
    df = df.assign(League=league)
    frame_list.append(df)

# Examples for callers:
#   - Get the E0 dataframe: frame_list[0]
#   - Get the E1 dataframe: frame_list[1]
#   - Recommended direct access by mapping:
#       frames_by_league = {df['League'].iat[0]: df for df in frame_list}
#       frames_by_league['E2']
# Avoid using print(frame_list) for large lists — inspect individual frames instead.
```

### Multiple Seasons


```python
for season in range(20, 26):
    print(str(season) + str(season+1))
```

```python
# creating a list of leagues and seasons (multi-season load)
# NOTE: frame_list is appended in outer-loop (league) then inner-loop (season).
# Order mapping: index = (league_index * n_seasons) + season_index
#   e.g. with 3 leagues and 5 seasons: indices 0-4 -> league0 seasons, 5-9 -> league1 seasons, etc.
leagues = ["E0", "E1", "E2"]
frame_list = []  # create an empty list to store the dataframes

root = "https://www.football-data.co.uk/mmz4281/"
seasons = [f"{s}{s+1}" for s in range(21, 26)]  # ['2122','2223','2324','2425','2526']

for league in leagues:
    for season_code in seasons:
        url = root + season_code + "/" + league + ".csv"
        try:
            df = pd.read_csv(url)
        except Exception as e:
            print(f"✗ Skipped: {url}  →  {e}")
            continue
        # label each frame so callers can select by League+Season reliably
        df = df.assign(League=league, Season=season_code)
        frame_list.append(df)
        print(f"✓ Loaded: {league} {season_code}")

# After loading, recommended access patterns:
# 1) By numeric index (if you know the order):
#      league_idx = 1  # index of 'E1' in `leagues`
#      season_idx = 2  # index of desired season in `seasons`
#      idx = league_idx * len(seasons) + season_idx
#      frame = frame_list[idx]
# 2) Prefer explicit lookup by (league,season):
#      frames_by_key = {(f['League'].iat[0], f['Season'].iat[0]): f for f in frame_list}
#      frames_by_key[('E1','2223')]
# 3) Or filter the combined DataFrame:
#      combined = pd.concat(frame_list, ignore_index=True)
#      combined[(combined['League']=='E1') & (combined['Season']=='2223')]

# Build combined and de-fragment for downstream work
if frame_list:
    combined = pd.concat(frame_list, ignore_index=True).copy()
    print(f"Loaded {len(frame_list)} frames; combined shape: {combined.shape}")
else:
    print("No frames loaded.")
```
