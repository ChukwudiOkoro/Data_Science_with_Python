# 1 Math operations


### 1.1 Operations in Columns

```python
# select a column and calculate total sum
df_exams ['math score'].sum()
```

```python
# count, mean, std max , and min
df_exams['math score'].count(), df_exams['math score'].mean(), df_exams['math score'].std(), df_exams['math score'].max(), df_exams['math score'].min()
```

```python
# easier calculation with .describe() for these values
df_exams.describe()
```

### 1.2  Operations in rows

```python
# calculating the sum in a row
df_exams ['math score']  + df_exams['reading score'] + df_exams['writing score']
```

```python
# calculating the average score  in a row and assign the values to a new column
df_exams['average']= (df_exams ['math score']  + df_exams['reading score'] + df_exams['writing score'])/3
```

# 2 Value count 

```python
# counting gender elements

# len function
len(df_exams['gender'])
# .count() method
df_exams['gender'].count()
```

```python


# Create sample data with NaN
df = pd.DataFrame({
    'gender': ['M', 'F', 'M', None, 'F']  # 1 missing value
})

# Using len()
print(len(df['gender']))           # Output: 5 (counts everything)

# Using .count()
print(df['gender'].count())        # Output: 4 (excludes NaN)
```
