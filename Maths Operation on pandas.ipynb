# 1 Math operations


### 1.1 Operations in Columns

```python
# select a column and calculate total sum
df_exams ['math score'].sum()
```

```python
# count, mean, std max , and min
df_exams['math score'].count(), df_exams['math score'].mean(), df_exams['math score'].std(), df_exams['math score'].max(), df_exams['math score'].min()
```

# 2 Value count 

```python
# counting gender elements

# len function
len(df_exams['gender'])
# .count() method
df_exams['gender'].count()
```

```python
#  count/ number of distinct/unique values ie M/F
df_exams['gender'].nunique()    
```

```python
# counting the number of unique values in a column
df_exams['gender'].value_counts()    
```

```python
# return the relative frequency(divide all values by the sum of values)
df_exams['gender'].value_counts(normalize=True)
```

```python
# Create sample data with NaN
df_NaN = pd.DataFrame({
    'gender': ['M', 'F', 'M', None, 'F']  # 1 missing value
})
```

```python
print(len(df_NaN['gender']))                     # Output: 5 (counts everything)
```

```python
# Using .count()
print(df_NaN['gender'].count())        # Output: 4 (excludes NaN)
```

```python
# calculating the sum in a row
df_exams['math score'] + df_exams['reading score'] + df_exams['writing score']
```

```python
#Sum across the three columns
df_sum = df_exams['math score'] + df_exams['reading score'] + df_exams['writing score']
print (df_sum)
```

```python
# counting "parental level of education" elements by value counts
df_exams['parental level of education'].value_counts()
```

```python
#return the relative frequency and sort the values in descending order
df_exams['parental level of education'].value_counts(normalize=True).sort_values(ascending=False) # normalize=True returns the relative frequency in decimal form, and sort_values(ascending=False) sorts the values in descending order.
```
